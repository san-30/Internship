```python
# 1. Rating

def webscrape_rating():
    rating1 = []
    rating1_element = driver.find_elements(By.XPATH, "//div[@class='_3LWZlK _1BLPMq']")
    time.sleep(2)
        
    for i in rating1_element:
        rating1.append(i.text)
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        
    return rating1
```

```python
# 2. Review Summary

def webscrape_review_summary():
    review_summary1 = []
    review_summary1_element = driver.find_elements(By.XPATH, "//p[@class='_2-N8zT']")
    time.sleep(2)
        
    for i in review_summary1_element:
        review_summary1.append(i.text)
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        
    return review_summary1
```

```python
# 3. Full Review

def webscrape_full_review():
    full_review1 = []
    full_review1_element = driver.find_elements(By.XPATH, "//div[@class='t-ZTKy']")
    time.sleep(2)
        
    for i in full_review1_element:
        full_review1.append(i.text.replace("\n"," "))
        driver.execute_script("window.scrollBy(0,document.body.scrollHeight0)")
        
    return full_review1
```

```python
# You have to scrape this data for first 100 reviews.

rating_list = []
review_summary_list = []
full_review_list = []

length = len(rating_list)

while(length <= 90):
    driver.refresh()
    rating_list.extend(webscrape_rating())
    
    review_summary_list.extend(webscrape_review_summary())
    
    full_review_list.extend(webscrape_full_review())
    
    time.sleep(3)
    
    length = len(rating_list)
    
    button_next = driver.find_element(By.XPATH, "//a[@class='_1LKTO3']")
    button_next.click()
    
len(rating_list)
```
